```python
from pyspark import SparkConf, SparkContext
from pyspark.sql.types import *
lines = sc.textFile("/resources/data/adhar/adhar.csv")


header = lines.filter(lambda l: "Enrolment Agency" in l)
noheader = lines.subtract(header)

def id_state(line):
    fields=line.split(',')
    return (fields[2],int(fields[8]))

def id_er(line):
    fields=line.split(',')
    return (fields[1],int(fields[8]))

def id_district(line):
    fields=line.split(',')
    return (fields[3],int(fields[8]))


#total id by state
totalidbyst=noheader.map(id_state).reduceByKey(lambda x,y:x+y)
totalidbyst.collect()
```

```python
#total id by Enrolment Agency
totalidbyea=noheader.map(id_er).reduceByKey(lambda x,y:x+y)
totalidbyea.collect()
```

```python
#total id by Enrolment Agency
rdd=noheader.map(id_district)
totalidbydis=rdd.mapValues(lambda x:(x,1)).reduceByKey(lambda x,y:(x[0]+y[0],x[1]+y[1]))
totalidbydis.collect()
```

```python
flipped=totalidbydis.map(lambda x:(x[1],x[0]))
sortedbydis=flipped.sortByKey(ascending=False)
topdis=sortedbydis.collect()
```
